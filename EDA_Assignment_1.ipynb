```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Create meshgrid
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3,  3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10,  8))

# Plot 1: pcolormesh()
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0,  0].set_title('pcolormesh()')

# Plot 2: contourf()
co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25,  1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0,  1].set_title('contourf()')

# Plot 3: imshow() with LogNorm()
pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma', norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1,  0], extend='both')
axs[1,  0].set_title('imshow() with LogNorm()')

# Plot 4: scatter()
pc = axs[1, 1].scatter(tips['total_bill'], tips['tip'], c=tips['size'], cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1,  1].set_title('scatter()')

# Set overall title
fig.suptitle("Visualization of Tips Dataset")

# Show the plot
plt.show()
```

```python
# Create a box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill Distribution by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

A box plot shows the distribution of quantitative data, highlighting the median, quartiles, and potential outliers. Here’s how to create a basic box plot for the “total_bill” column in the “tips” dataset:

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", palette="Set2", width=0.6)
plt.title("Grouped Box Plot: Total Bill Distribution by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

Grouped Box Plot: To compare the distribution of total bills across different days, you can create a grouped box plot. This will show separate boxes for each day side by side.

```python
sns.boxplot(x="total_bill", y="day", data=tips, hue="sex", palette="Set2", width=0.6, orient="h")
plt.title("Horizontal Box Plot: Total Bill Distribution by Day and Gender")
plt.xlabel("Total Bill ($)")
plt.ylabel("Day of the Week")
plt.show()
```

Horizontal Box Plot: If you prefer a horizontal orientation, you can switch the x and y axes:
plt.title("Horizontal Box Plot: Total Bill Distribution by Day and Gender"): Sets the title of the plot.
plt.xlabel("Total Bill ($)"): Adds a label to the x-axis.
plt.ylabel("Day of the Week"): Adds a label to the y-axis.

```python
sns.violinplot(x="day", y="tip", data=tips, palette="Set2")
plt.title("Tip Distribution by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Tip ($)")
plt.show()
```

A violin plot combines a box plot with a kernel density estimation (KDE) to show the distribution of data. Here’s an example using the “tip” and “total_bill” columns:

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex", palette="Set2", split=True)
plt.title("Grouped Violin Plot: Tip Distribution by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Tip ($)")
plt.show()
```

Grouped Violin Plot: To compare the distribution of tips across different days, you can create a grouped violin plot. This will show separate violins for each day side by side.

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex", palette="Set2")
plt.title("Stacked Violin Plot: Tip Distribution by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Tip ($)")
plt.show()
```

Stacked Violin Plot: If you want to stack the violins instead of splitting them, you can remove the split=True parameter:

```python
sns.jointplot(data=tips, x="tip", y="total_bill", palette="Set2")
plt.show()
```

A joint plot displays the relationship between two variables along with their individual distributions. Here’s how to create a scatter plot with marginal histograms:

```python
sns.jointplot(data=tips, x="tip", y="total_bill", hue="day", kind="kde", palette="Set2")
plt.show()
```

Kernel Density Estimate (KDE) Plot: This plot shows both bivariate and univariate KDEs. It provides a smooth representation of the data density.

```python
sns.jointplot(data=tips, x="tip", y="total_bill", kind="reg", palette="Set2")
plt.show()
```

Regression Plot with KDE Curves: You can add a linear regression fit along with univariate KDE curves.

```python
sns.jointplot(data=tips, x="tip", y="total_bill", kind="hex", palette="Set2")
plt.show()
```

Hexbin Plot: A hexbin plot divides the data space into hexagonal bins and shows the count of points within each bin.

```python
# Extracting relevant columns from the dataset
data1 = tips['total_bill']
data2 = tips['tip']
data3 = tips['size']
data4 = tips['tip'] / tips['total_bill']  # Calculating tip percentage

# Creating the plot
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(data1, 'o', label='total_bill')
ax.plot(data2, 'd', label='tip')
ax.plot(data3, 'v', label='size')
ax.plot(data4, 's', label='tip_percentage')
ax.legend()

plt.show()
```

 the code loads the "tips" dataset, extracts specific columns of interest, creates a plot with those columns, and displays the plot with a legend.

```python
# Extract relevant columns
total_bill = tips['total_bill']
tip = tips['tip']
size = tips['size']
tip_percentage = tips['tip'] / tips['total_bill']

# Single subplot
fig, ax = plt.subplots()
ax.hist(total_bill, bins=20, color='skyblue', edgecolor='black')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

```python
# Multiple subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].scatter(total_bill, tip, color='orange')
axs[0].set_xlabel('Total Bill')
axs[0].set_ylabel('Tip')
axs[1].scatter(total_bill, size, color='green')
axs[1].set_xlabel('Total Bill')
axs[1].set_ylabel('Size')
plt.tight_layout()
plt.show()
```

```python
# Multiple subplots (2 rows, 1 column)
fig, axs = plt.subplots(2, 1, figsize=(6, 8))
axs[0].hist(tip, bins=15, color='purple', edgecolor='black')
axs[0].set_xlabel('Tip')
axs[0].set_ylabel('Frequency')
axs[1].scatter(size, tip_percentage, color='red')
axs[1].set_xlabel('Size')
axs[1].set_ylabel('Tip Percentage')
plt.tight_layout()
plt.show()
```

```python
# Multiple subplots (2x2 grid)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].hist(total_bill, bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_xlabel('Total Bill')
axs[0, 0].set_ylabel('Frequency')
axs[0, 1].scatter(tip, tip_percentage, color='orange')
axs[0, 1].set_xlabel('Tip')
axs[0, 1].set_ylabel('Tip Percentage')
axs[1, 0].scatter(total_bill, size, color='green')
axs[1, 0].set_xlabel('Total Bill')
axs[1, 0].set_ylabel('Size')
axs[1, 1].hist(size, bins=10, color='purple', edgecolor='black')
axs[1, 1].set_xlabel('Size')
axs[1, 1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
